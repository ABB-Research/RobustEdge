# 00 — Dataset overview

Discover the phase-aware campaign structure and inspect run metadata.

```python
from pathlib import Path
from robustedge.data import DatasetIndex

DATA_ROOT = Path('../data/raw/logs')
if not DATA_ROOT.exists():
    DATA_ROOT = Path('../tests/fixtures/minimal_dataset')
index = DatasetIndex.from_root(DATA_ROOT)
manifest = index.to_frame()
print(f'Discovered {len(manifest)} runs under {DATA_ROOT.resolve()}')
manifest.head()
```

```python
manifest.groupby(['phase','perturbation_family','severity','attack_duration'], dropna=False).size().reset_index(name='n_runs')
```
